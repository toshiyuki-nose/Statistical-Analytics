## Sample Dataset

The dataset contains fictional countries and regions. It resembles socioeconomic and demographic country data, but it does not represent real countries.

Each row is one **observation**. Each column is one **variable**.

```python
from pathlib import Path

import pandas as pd
```

```python
# Find the repository root from either the repository root or notebooks/.
current_dir = Path.cwd().resolve()
candidates = [current_dir, *current_dir.parents]
project_root = next((path for path in candidates if (path / 'data').is_dir()), None)

if project_root is None:
    raise FileNotFoundError('Could not find the project root containing the data directory.')

data_path = project_root / 'data' / '01_data_and_variables.csv'
print(f'Project root: {project_root}')
print(f'Data path: {data_path}')
```

```python
df = pd.read_csv(data_path)
df.head()
```

## Inspect the Dataset

The shape shows the number of observations and variables. The data types provide an initial clue about whether a variable is stored as text or as a number.

```python
print(f'Observations: {df.shape[0]}')
print(f'Variables: {df.shape[1]}')
print('\nColumn names:')
print(df.columns.tolist())
print('\nData types:')
print(df.dtypes)
```

## Variable Classification

| Variable | Statistical type | Measurement scale | Meaning |
|---|---|---|---|
| `country_id` | Categorical | Nominal | Fictional country identifier |
| `region` | Categorical | Nominal | Geographic group |
| `income_group` | Categorical | Ordinal | Ordered income category |
| `population_million` | Quantitative, continuous | Ratio | Population in millions |
| `gdp_per_capita_usd` | Quantitative, continuous | Ratio | GDP per person in USD |
| `life_expectancy` | Quantitative, continuous | Ratio | Expected years of life |
| `urbanization_rate` | Quantitative, continuous | Ratio | Percentage living in urban areas |

```python
categorical_cols = ['country_id', 'region', 'income_group']
quantitative_cols = [
    'population_million',
    'gdp_per_capita_usd',
    'life_expectancy',
    'urbanization_rate',
]

print('Categorical variables:')
print(df[categorical_cols].head())

print('\nQuantitative variables:')
print(df[quantitative_cols].head())
```

## Simple Checks

Before analysis, check missing values and the categories contained in categorical variables.

```python
print('Missing values by variable:')
print(df.isna().sum())

print('\nIncome groups:')
print(df['income_group'].value_counts())
```

## Summary

- Rows represent observations, and columns represent variables.
- Categorical variables describe groups or labels.
- Quantitative variables express measurable amounts.
- Measurement scales determine which statistical operations are meaningful.
- Data types in pandas and statistical variable types are related, but they are not identical.
